```python
################################################################
##################### import libraries  ########################
################################################################

import sys,time





################################################################
##################### define All rooms  ########################
################################################################

game_room = {
    "name": "game room",
    "type": "room",
    "description": "The room is large and bare. You can smell the mold tainting the walls.",
}

bed1 = {
    "name": "bedroom 1",
    "type": "room",
    "description": "The room is dark, with thick curtains making the room feel much heavier. It feels as if someone or something is watching you.",
    #heavy curtains blocking out any light. 
}

bed2 = {
    "name": "bedroom 2",
    "type": "room",
    "description": "A cozy bedroom with a welcoming atmosphere, but as you look closer, you notice small scratches and bloodstains on the walls and floor. Amidst those scratches you can read  - OMG i forgot to submit 4 labs!. Below that you can read something that resembles 'Fred'",
}

living_room = {
    "name": "living room",
    "type": "room",
    "description": "You see yourself on a large room filled with the smell of raw meat that went bad. You gain a sense of hope seeing a ray of light shining through the window. Some old Python books sit destroyed on the floor.",
}

outside1 = {
  "name": "outside", 
  "description": " ",
}

################################################################
############## defines ALL interconnections  ###################
################################################################

door_a = {
    "name": "door a",
    "type": "door",
    "description": "",
}

door_b = {
    "name": "door b",
    "type": "door",
     "description": "",
}

door_c = {
    "name": "door c",
    "type": "door",
     "description": "",
}

window_a = {
    "name": "window a",
    "type": "window",
     "description": "You approach the strange window. Blue lightnings sizzle from the edges! Strange letters begin to form on the wooden frame! Atop it's written: Choose wisely... ",
}

################################################################
############# defines ALL keys/items required ##################
################################################################

key_a = {
    "name": "key for door a",
    "type": "key",
    "target": door_a,
     "description": "",
}

key_b = {
    "name": "key for door b",
    "type": "key",
    "target": door_b,
     "description": "",
}

key_c = {
    "name": "key for door c",
    "type": "key",
    "target": door_c,
    "description": "",
}

#key_d = {
#    "name": "key for door d",
#    "type": "key",
#    "target": door_d,
#     "description": "",
#}

################################################################
############## define items inside rooms  ######################
################################################################

couch = {
    "name": "couch",
    "type": "furniture",
    "description": "You look at the couch pushed into the corner of the room. You can see it has had better days. You wonder how you woke up there...",
}

piano = {
    "name": "piano",
    "type": "furniture",
    "description": "A broken down piano. For the dust on top of it, it hasn't been played for decades.",
}

queenbed = {
    "name": "queen bed",
    "type": "furniture",
    "description": "A big bed that looks magestic but as you approach, you see something that resembles HUMAN bones on the pillow.",
}

doublebed = {
    "name": "double bed",
    "type": "furniture",
    "description": "A simple double bed. It doesn't smell good!",
}

dresser = {
    "name": "dresser",
    "type": "furniture",
    "description": "A small dresser with no particular signs. You notice however a small crown standing on top of it!",
}

diningtable = {
    "name": "dining table",
    "type": "furniture",
    "description": " Table is set for dinner. Flies cover the dishes. Maybe someone forgot to eat it while taking some intensive Data course!",
}

################################################################
############## define all special items ########################
################################################################

all_rooms = [game_room, bed1, bed2, living_room, outside1]

all_doors = [door_a,door_b,door_c,window_a]

################################################################
############## define relations between objects ################
################################################################

object_relations = {
    "game room": [couch, piano, door_a],
    "piano": [key_a],
    "outside1": [window_a],
    "door a": [game_room, bed1], 
    "bedroom 1":[queenbed,door_c,door_b],
    "queen bed":[key_b],
    "door c":[bed1,living_room],
    "door b":[bed1,bed2],
    "window a":[living_room,outside1],
    "bedroom 2":[doublebed,dresser,door_b],
    "double bed":[key_c],
#    "dresser":[key_d],
    "living room":[diningtable,window_a],
}

################################################################
############ Standard (fixed) game start point #################
################################################################

# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_room": game_room, 
    "keys_collected": [],
    "target_room": outside1
}
```

```python
def start_game(debug=False):
    """
    Start the game
    for debugging, one can pass debug=True as argument
    """
    if debug == True:
        game_state["current_room"] = eval(input("Pick one from game_room, bed1, bed2, living_room: ").replace("'",'"'))
        game_state["keys_collected"] = [key_a, key_b, key_c]
        
    choose_character()
    slow_type("You wake up on a couch and find yourself in a strange house. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of the house, NOW!")
    play_room(game_state["current_room"])

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        slow_type("\nThe whole room shakes as the magic window opens... A strong gust of wind forces you through the window into the lust green openspace of outside. You can smell all the flowers and the big Pine tree sitting next to the haunted house you just escaped from!")
        print("Congrats! You escaped the house!")
    else:
        #print("You are now in " + room["name"])
        intended_action = input("What would you like to do? Type 'explore' or 'examine'?").strip().lower()
        if intended_action == "explore":
            explore_room(room)
            play_room(room)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine?").strip().lower())
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
            play_room(room)
        print("\n\n")

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    slow_type("\nYou explore the " + room["name"] + ". " + room["description"] + " You find " + ", ".join(items) + "\n")

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None

    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = ""
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have.\n"
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the key."
            elif(item["type"] == "window"):
                slow_type("\nBelow runic symbols compress into words in front of your eyes. They spell:")
                [print(f"{i + 1} - {team}") for i, team in enumerate(group_names)]
                while True:
                    try:
                        choice = int(input("Which name will you press? :"))
                    except:
                        slow_type("\nWith the fear you're hand shakes... the button you barely touched shimmers but nothing happens. Will you try again?")
                    else:
                        if choice in range(1,7):
                            if choice == 2:
                                next_room = get_next_room_of_door(item, current_room)
                                play_room(next_room)
                                return False
                            else:
                                slow_type("As you press the button, you feel you're whole arm going numb. You understand you made the wrong choice!")
                        else:
                            slow_type("\nWith the fear you're hand shakes... the button you barely touched shimmers but nothing happens. Will you try again?")
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "\n" + item["description"] + ". \nYou find the " + item_found["name"] + ".\n" 
                else:
                    output += "\n" + item["description"] + "\nThere isn't anything interesting about it.\n"
            slow_type(output)
            break

    if(output is None):
        print("The item you requested is not found in the current room.")
    
    if(next_room and input("Do you want to go to the next room? Enter 'yes' or 'no'").strip() == 'yes'):
        play_room(next_room)
    else:
        play_room(current_room)

def choose_character():
    """
    Allow the player to choose a character at the start of the game.
    """
    characters = ["Princess", "Dragon", "Knight", "Wizard", "Thief"]
    print("Choose your character:")
    for i, char in enumerate(characters):
        print(f"{i + 1}. {char}")
    choice = int(input("Enter the number of your chosen character: ").strip()) - 1
    if 0 <= choice < len(characters):
        slow_type(f"You chose the {characters[choice]}!")
        character=characters[choice]
    else:
        print("Invalid choice, starting as a Knight by default.")
        choice = 2
        character="Knight"

def slow_type(str):
    for c in str + "\n":
        sys.stdout.write(c)
        sys.stdout.flush()
        time.sleep(3./90)


       
```
